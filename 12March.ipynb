```python
import scipy.stats as stats
import numpy as np

# Given data
sample_mean = 50
sample_std = 5
sample_size = 30  # Assuming a sample size of 30 for the confidence interval

# 95% confidence level
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Margin of error
margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

# Confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(f"95% Confidence Interval: {confidence_interval}")
```

```python
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([35, 30, 25, 20, 10, 20])  # Replace with actual observed frequencies
# Expected frequencies
expected = np.array([20, 20, 20, 10, 10, 20]) * (sum(observed) / 100)

# Chi-square goodness of fit test
chi2_stat, p_value = chisquare(observed, f_exp=expected)
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_value}")

# Significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The distribution of M&Ms does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of M&Ms matches the expected distribution.")
```

```python
#6 
import matplotlib.pyplot as plt

# Chi-square distribution with 10 degrees of freedom
df = 10
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

plt.plot(x, y, label='Chi-square distribution with 10 df')
plt.fill_between(x, y, where=(x >= 15), color='red', alpha=0.3, label='Chi-square stat >= 15')
plt.title('Chi-square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-square value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
#8
# Observed frequencies
observed = np.array([45, 55])  # 45 tails, 55 heads
# Expected frequencies
expected = np.array([50, 50])

# Chi-square goodness of fit test
chi2_stat, p_value = chisquare(observed, f_exp=expected)
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_value}")

# Significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The coin is biased.")
else:
    print("Fail to reject the null hypothesis: The coin is not biased.")
```

```python
#11
# Given data
sample_mean = 72
sample_std = 10
sample_size = 30
population_mean = 70

# T-statistic
t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = stats.t.sf(np.abs(t_statistic), df=sample_size-1) * 2

# Significance level
alpha = 0.05
print(f"T-statistic: {t_statistic}, P-value: {p_value}")
if p_value < alpha:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: The population mean is not significantly different from 70.")
```

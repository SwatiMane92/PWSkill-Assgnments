```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.utils import resample
import matplotlib.pyplot as plt

# Q1. What is an ensemble technique in machine learning?
print("Q1. What is an ensemble technique in machine learning?")
print("An ensemble technique in machine learning combines the predictions of multiple models to improve overall performance and accuracy.")

# Q2. Why are ensemble techniques used in machine learning?
print("\nQ2. Why are ensemble techniques used in machine learning?")
print("Ensemble techniques are used to improve model performance, reduce variance and bias, and achieve better generalization by combining the strengths of multiple models.")

# Q3. What is bagging?
print("\nQ3. What is bagging?")
print("Bagging, or Bootstrap Aggregating, is an ensemble technique that trains multiple models on different subsets of the training data (created through bootstrapping) and combines their predictions, usually by averaging or voting.")

# Q4. What is boosting?
print("\nQ4. What is boosting?")
print("Boosting is an ensemble technique that trains multiple models sequentially, where each model corrects the errors made by the previous ones. The predictions are combined in a weighted manner to improve overall performance.")

# Q5. What are the benefits of using ensemble techniques?
print("\nQ5. What are the benefits of using ensemble techniques?")
print("The benefits of ensemble techniques include improved model accuracy, reduced risk of overfitting, increased robustness, and better handling of complex datasets by leveraging the strengths of multiple models.")

# Q6. Are ensemble techniques always better than individual models?
print("\nQ6. Are ensemble techniques always better than individual models?")
print("Not always. Ensemble techniques can improve performance, but they also require more computation and may not always outperform individual models, especially if the base models are poor or if the data is not complex enough.")

# Q7. How is the confidence interval calculated using bootstrap?
print("\nQ7. How is the confidence interval calculated using bootstrap?")
print("The confidence interval using bootstrap is calculated by repeatedly sampling with replacement from the original dataset, computing the statistic of interest (e.g., mean) for each sample, and then determining the interval within which a certain percentage of these statistics fall.")

# Q8. How does bootstrap work and what are the steps involved in bootstrap?
print("\nQ8. How does bootstrap work and what are the steps involved in bootstrap?")
print("Bootstrap works by creating multiple samples from the original dataset through random sampling with replacement. The steps are:")
print("1. Draw multiple bootstrap samples (with replacement) from the original dataset.")
print("2. Calculate the statistic of interest (e.g., mean) for each bootstrap sample.")
print("3. Analyze the distribution of these statistics to estimate the confidence interval or other properties.")

# Q9. Bootstrap confidence interval example
print("\nQ9. Bootstrap confidence interval example")

# Parameters
np.random.seed(42)  # For reproducibility
sample_size = 50
mean_height = 15
std_dev = 2
n_iterations = 1000
confidence_level = 0.95

# Generate original sample
original_sample = np.random.normal(mean_height, std_dev, sample_size)

# Bootstrap procedure
bootstrap_means = []
for _ in range(n_iterations):
    bootstrap_sample = resample(original_sample, replace=True)
    bootstrap_means.append(np.mean(bootstrap_sample))

# Calculate the confidence interval
lower_bound = np.percentile(bootstrap_means, (1 - confidence_level) / 2 * 100)
upper_bound = np.percentile(bootstrap_means, (1 + confidence_level) / 2 * 100)

# Print results
print(f"Bootstrap 95% Confidence Interval for the mean height: ({lower_bound:.2f}, {upper_bound:.2f}) meters")

# Plot histogram of bootstrap means
plt.hist(bootstrap_means, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(x=lower_bound, color='r', linestyle='--', label=f'Lower bound ({lower_bound:.2f})')
plt.axvline(x=upper_bound, color='g', linestyle='--', label=f'Upper bound ({upper_bound:.2f})')
plt.title('Histogram of Bootstrap Means')
plt.xlabel('Mean Height')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

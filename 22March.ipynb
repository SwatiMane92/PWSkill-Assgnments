```python
# Import necessary libraries
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Q1: Pearson correlation coefficient between study time and exam scores

# Sample data: Study time (hours) and Exam scores (out of 100)
study_time = np.array([2, 3, 5, 7, 8, 9, 10])
exam_scores = np.array([55, 60, 70, 80, 85, 90, 95])

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(study_time, exam_scores)
pearson_corr

# Interpretation
pearson_corr_interpretation = """
The Pearson correlation coefficient between study time and exam scores is {:.2f}.
A value close to 1 indicates a strong positive linear relationship, meaning as study time increases, exam scores tend to increase as well.
""".format(pearson_corr)

# Q2: Spearman's rank correlation between sleep and job satisfaction

# Sample data: Hours of sleep and Job satisfaction (scale of 1 to 10)
sleep_hours = np.array([6, 7, 5, 8, 7, 6, 9])
job_satisfaction = np.array([4, 5, 3, 7, 6, 5, 8])

# Calculate Spearman's rank correlation coefficient
spearman_corr, _ = spearmanr(sleep_hours, job_satisfaction)
spearman_corr

# Interpretation
spearman_corr_interpretation = """
The Spearman's rank correlation coefficient between hours of sleep and job satisfaction is {:.2f}.
A value close to 1 indicates a strong positive monotonic relationship, meaning as sleep increases, job satisfaction tends to increase.
""".format(spearman_corr)

# Q3: Pearson and Spearman correlations between hours of exercise and BMI

# Sample data: Hours of exercise per week and BMI
exercise_hours = np.array([1, 2, 3, 4, 5, 6, 7])
bmi = np.array([25, 24, 23, 22, 21, 20, 19])

# Calculate Pearson correlation coefficient
pearson_corr_exercise_bmi, _ = pearsonr(exercise_hours, bmi)

# Calculate Spearman's rank correlation coefficient
spearman_corr_exercise_bmi, _ = spearmanr(exercise_hours, bmi)

# Interpretation
pearson_corr_exercise_bmi_interpretation = """
The Pearson correlation coefficient between hours of exercise and BMI is {:.2f}.
A negative value indicates that as hours of exercise increase, BMI tends to decrease, which is expected.
""".format(pearson_corr_exercise_bmi)

spearman_corr_exercise_bmi_interpretation = """
The Spearman's rank correlation coefficient between hours of exercise and BMI is {:.2f}.
A negative value indicates a negative monotonic relationship, which complements the Pearson result.
""".format(spearman_corr_exercise_bmi)

# Q4: Pearson correlation coefficient between TV watching time and physical activity level

# Sample data: Hours spent watching TV per day and Physical activity level (scale of 1 to 10)
tv_hours = np.array([2, 3, 4, 5, 6, 7, 8])
physical_activity = np.array([7, 6, 5, 4, 3, 2, 1])

# Calculate Pearson correlation coefficient
pearson_corr_tv_activity, _ = pearsonr(tv_hours, physical_activity)

# Interpretation
pearson_corr_tv_activity_interpretation = """
The Pearson correlation coefficient between TV watching time and physical activity level is {:.2f}.
A negative value indicates that as TV watching time increases, physical activity level tends to decrease.
""".format(pearson_corr_tv_activity)

# Q5: Analyzing the relationship between age and preference for a brand

# Sample data: Age and Preference score (scale of 1 to 10)
age = np.array([22, 25, 30, 35, 40, 45, 50])
preference_score = np.array([6, 7, 8, 5, 4, 3, 2])

# Calculate Pearson correlation coefficient
pearson_corr_age_preference, _ = pearsonr(age, preference_score)

# Interpretation
pearson_corr_age_preference_interpretation = """
The Pearson correlation coefficient between age and preference for the brand is {:.2f}.
A negative value suggests that as age increases, the preference for the brand tends to decrease.
""".format(pearson_corr_age_preference)

# Q6: Analyzing the relationship between sales calls and sales made

# Sample data: Number of sales calls per day and Number of sales per week
sales_calls = np.array([10, 20, 30, 40, 50, 60, 70])
sales_made = np.array([5, 10, 15, 20, 25, 30, 35])

# Calculate Pearson correlation coefficient
pearson_corr_sales_calls_sales, _ = pearsonr(sales_calls, sales_made)

# Interpretation
pearson_corr_sales_calls_sales_interpretation = """
The Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week is {:.2f}.
A positive value indicates a strong positive linear relationship, meaning more sales calls are associated with more sales.
""".format(pearson_corr_sales_calls_sales)

# Display results
pearson_corr, spearman_corr, pearson_corr_exercise_bmi, spearman_corr_exercise_bmi, pearson_corr_tv_activity, pearson_corr_age_preference, pearson_corr_sales_calls_sales
```

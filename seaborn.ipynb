```python
# Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

**Five plots in Seaborn and their uses**:

1. **Scatter Plot** (`sns.scatterplot`):
   - **Use**: Visualizes the relationship between two continuous variables.

2. **Line Plot** (`sns.lineplot`):
   - **Use**: Shows trends over time or other ordered categories.

3. **Box Plot** (`sns.boxplot`):
   - **Use**: Displays the distribution of a continuous variable and identifies outliers.

4. **Histogram** (`sns.histplot`):
   - **Use**: Shows the distribution of a single continuous variable.

5. **Pair Plot** (`sns.pairplot`):
   - **Use**: Provides pairwise plots of all numeric variables in a dataset to understand relationships between them.
```

```python
# Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('Line Plot of Signal vs Timepoint')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event and Region')
plt.show()
```

```python
# Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
plt.figure(figsize=(14, 6))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
# Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=30)
plt.title('Histogram of Price by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset("iris")

# Plot the pair plot
sns.pairplot(data=iris, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python
# Q6: Use the "flights" dataset from seaborn to plot a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
flights = sns.load_dataset("flights").pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data=flights, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Heatmap of Number of Passengers')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

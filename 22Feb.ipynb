```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the YouTube channel's video page
url = "https://www.youtube.com/@PW-Foundation/videos"

# Send a request to the URL
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Function to extract video information
def extract_video_info(soup, num_videos=5):
    video_data = []
    
    # Find all video elements
    video_elements = soup.find_all('ytd-grid-video-renderer', limit=num_videos)
    
    for video in video_elements:
        # Extract video URL
        video_url = "https://www.youtube.com" + video.find('a', {'id': 'thumbnail'})['href']
        
        # Extract thumbnail URL
        thumbnail_url = video.find('img', {'id': 'img'})['src']
        
        # Extract title
        title = video.find('a', {'id': 'video-title'})['title']
        
        # Extract number of views
        views = video.find('span', {'class': 'style-scope ytd-grid-video-renderer'}).text
        
        # Extract time of posting
        time_posted = video.find('div', {'id': 'metadata-line'}).find_all('span')[1].text
        
        video_data.append({
            'Video URL': video_url,
            'Thumbnail URL': thumbnail_url,
            'Title': title,
            'Views': views,
            'Time Posted': time_posted
        })
    
    return video_data

# Extract video information
video_info = extract_video_info(soup)

# Save data to CSV
csv_file = 'youtube_video_data.csv'
csv_columns = ['Video URL', 'Thumbnail URL', 'Title', 'Views', 'Time Posted']

with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
    writer.writeheader()
    writer.writerows(video_info)

print(f"Data has been written to {csv_file}")
```

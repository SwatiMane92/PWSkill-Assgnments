```python
# Import necessary libraries
import numpy as np
import pandas as pd

# Q1. What is Bayes' theorem?
print("Q1. What is Bayes' theorem?")
print("Bayes' theorem is a fundamental theorem in probability theory that describes how to update the probability of a hypothesis based on new evidence.")
print("It provides a way to calculate the posterior probability of an event based on prior knowledge and evidence.")

# Q2. What is the formula for Bayes' theorem?
print("\nQ2. What is the formula for Bayes' theorem?")
print("The formula for Bayes' theorem is:")
print("P(A|B) = (P(B|A) * P(A)) / P(B)")
print("where:")
print("P(A|B) = Posterior probability of A given B")
print("P(B|A) = Likelihood of B given A")
print("P(A) = Prior probability of A")
print("P(B) = Prior probability of B")

# Q3. How is Bayes' theorem used in practice?
print("\nQ3. How is Bayes' theorem used in practice?")
print("Bayes' theorem is used in various applications such as:")
print("- Spam filtering in email services")
print("- Medical diagnostics")
print("- Risk assessment in finance")
print("- Predictive text input")

# Q4. What is the relationship between Bayes' theorem and conditional probability?
print("\nQ4. What is the relationship between Bayes' theorem and conditional probability?")
print("Bayes' theorem is fundamentally based on the concept of conditional probability.")
print("It calculates the conditional probability of an event by updating prior probabilities with new evidence.")

# Q5. How do you choose which type of Naive Bayes classifier to use for any given problem?
print("\nQ5. How do you choose which type of Naive Bayes classifier to use for any given problem?")
print("The choice of Naive Bayes classifier depends on the nature of the features:")
print("- Gaussian Naive Bayes: Used when features are continuous and normally distributed.")
print("- Multinomial Naive Bayes: Used when features are counts or frequencies (e.g., text classification).")
print("- Bernoulli Naive Bayes: Used when features are binary or boolean (e.g., presence/absence of a feature).")

# Q6. Assignment: Naive Bayes Classification
print("\nQ6. Assignment: Naive Bayes Classification")
# Define the frequency table for each class
data = {
    'Class': ['A', 'B'],
    'X1=1': [3, 2],
    'X1=2': [3, 2],
    'X1=3': [4, 1],
    'X2=1': [4, 2],
    'X2=2': [3, 2],
    'X2=3': [3, 2],
    'X2=4': [3, 3]
}
df = pd.DataFrame(data)

# Calculate prior probabilities (assuming equal priors)
prior_prob_A = 0.5
prior_prob_B = 0.5

# Calculate likelihoods
def calculate_likelihood(class_name, x1, x2):
    x1_col = f'X1={x1}'
    x2_col = f'X2={x2}'
    
    likelihood_x1 = df.loc[df['Class'] == class_name, x1_col].values[0] / df.loc[df['Class'] == class_name, x1_col].sum()
    likelihood_x2 = df.loc[df['Class'] == class_name, x2_col].values[0] / df.loc[df['Class'] == class_name, x2_col].sum()
    
    return likelihood_x1 * likelihood_x2

# New instance features
x1_new = 3
x2_new = 4

# Calculate posterior probabilities
likelihood_A = calculate_likelihood('A', x1_new, x2_new)
likelihood_B = calculate_likelihood('B', x1_new, x2_new)

posterior_A = likelihood_A * prior_prob_A
posterior_B = likelihood_B * prior_prob_B

# Predict class
predicted_class = 'A' if posterior_A > posterior_B else 'B'

print(f"Posterior probability for class A: {posterior_A:.4f}")
print(f"Posterior probability for class B: {posterior_B:.4f}")
print(f"The predicted class for the new instance (X1={x1_new}, X2={x2_new}) is: {predicted_class}")
```
